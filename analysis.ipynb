```python
import pandas as pd

File = 'swiggy.csv'

df = pd.read_csv(File)

df.head(5)
```

```python
# How many rows?
len(df)
# What columns exist?
df.columns

# Any missing values?
df.isnull().sum()

# Data types correct? (date, numbers)
df.info()
```

```python
# Top 5 Restaurants by rating

df.sort_values(by='Avg ratings', ascending=False).head(5)
```

```python
# Top 5 by popularity
df.sort_values(by='Total ratings', ascending=False).head(5)
```

```python
# Avg delivery time per city
df.groupby('City')['Delivery time'].mean()
```

```python
# sort by avg rating , total ratings values
best_restaurants = df.sort_values(by=['Avg ratings', 'Total ratings'], ascending=[False, False]).head(5)
best_restaurants
```

```python
# Top 5 city by avg rating and total ratings

city_analysis = df.groupby('City').agg({'Avg ratings' : 'mean', 'Total ratings': 'mean'}).reset_index()

city_analysis = city_analysis.sort_values(by=['Avg ratings', 'Total ratings'], ascending=[False , False])
city_analysis.head(5)
```

```python
# Top Restaurant by City
top_restaurant = df.groupby(['City', 'Restaurant']).agg({'Avg ratings': 'mean', 'Total ratings': 'sum'}).reset_index()
top_restaurant = top_restaurant.sort_values(by=['Avg ratings', 'Total ratings'], ascending=[False , False]).groupby('City').head(1)
top_restaurant
```

```python
df.groupby('Restaurant')['Total ratings'].sum().sort_values(ascending=False)
```

```python
# Unique Restaurants count by city
unique_restaurants = df.groupby('City')['Restaurant'].nunique()
unique_restaurants
```

```python
# Top 3 restaurants per city
top_3_restaurant_by_city = df.groupby(['City', 'Restaurant']).agg({'Avg ratings': 'mean', 'Total ratings': 'sum'}).reset_index()
top_3_restaurant_by_city = top_3_restaurant_by_city.sort_values(by=['City', 'Avg ratings', 'Total ratings'] , ascending=[True,  False, False]).groupby('City').head(3)
top_3_restaurant_by_city
```
