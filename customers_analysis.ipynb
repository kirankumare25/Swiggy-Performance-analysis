```python
import pandas as pd

files = 'food_delivery_dataset.csv'

df = pd.read_csv(files)

df.head(5)
```

```python
# Average Order Frequency

average_order_frequency = len(df) / df['user_id'].nunique()
print(average_order_frequency)
```

```python
#  Repeat customer %
customers_orders = df.groupby('user_id')['order_id'].count().reset_index(name='orders')
repeat_customers = (len(customers_orders[customers_orders['orders'] > 1]) / customers_orders['user_id'].count())* 100
repeat_customers 
```

```python
# Top 10 users contribution to revenue by orders
top_contribution = df.groupby('user_id').agg({'order_id': 'count', 'final_amount': 'sum'}).sort_values(by='final_amount', ascending=False)
top_contribution.head(10)
```

```python
# Top 10% users contribution to revenue

user_revenue = df.groupby('user_id')['final_amount'].sum().sort_values(ascending=False)

top_10_percent = int(len(user_revenue) * 0.1)
top_users = user_revenue.head(top_10_percent)


top_contribution = top_users.sum() / user_revenue.sum() * 100

top_contribution
```

```python
# Revenue from repeat vs new users
user_orders = df.groupby('user_id')['order_id'].count()

repeat_customers = user_orders[user_orders > 1].index

df['user_type'] = df['user_id'].apply(lambda x: 'repeat' if x in repeat_customers else 'new')

# revenue_split = df.groupby('user_type')['final_amount'].sum()
# revenue_split_percentage = revenue_split / revenue_split.sum() * 100
# revenue_split_percentage 

revenue_split = df.groupby('user_type')['final_amount'].sum()
revenue_split

revenue_percentage = revenue_split / revenue_split.sum() * 100
revenue_percentage
```
